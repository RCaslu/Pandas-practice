#### Importe o banco de dados que está no seguinte [link](https://drive.google.com/file/d/1paQsoixucwIBvO8kQQmC6JD6lkrIam5Z/view?usp=sharing).

Este banco possui dados de estoque da empresa Pecista. As colunas existentes são:


*   Codigo: Codigo interno do SKU, os valores não devem se repetir;
*   Produto: Descrição do SKU;
*   Quantidade: Quatidade existente no estoque de um dado produto;
*   Altura: Altura do SKU em centímetros;
*   Largura: Largura do SKU em centímetros;
*   Produnfidade: Profundidade do SKU em centímetros.

```python
df = pd.read_csv("dados.csv", sep=";")
display(df)
```

- Utilize a coluna 'Codigo' como índice

```python
df = df.set_index('Codigo')
display(df)
```

- Quantos produtos distintos temos?

```python
qtd_produtos = df.shape[0]
print("O número de produtos diferentes é:",qtd_produtos)
```

- Existe algum produto com algum dado estranho?

```python
dado_nulo = df[df.isnull().any(axis=1)]
print(f"Um produto com valores incompletos foi idenficado:\nNome: {dado_nulo['Produto']}")
```

- Qual produto ocupa o maior volume de nosso estoque?

```python
df['Altura'] = df['Altura'].str.replace(',', '.').astype(float)
df['Largura'] = df['Largura'].str.replace(',', '.').astype(float)
df['Profundidade'] = df['Profundidade'].str.replace(',', '.').astype(float)
df['Quantidade'] = df['Quantidade'].str.replace(',', '.').astype(float)

df['Volume total'] = df['Altura'] * df['Largura'] * df['Profundidade'] * df['Quantidade']
id_maior_estoque = df['Volume total'].idxmax()
produto_maior_vol_estq = df.loc[id_maior_estoque]

print(f"O produto que ocupa maior volume de estoque é:\nNome: {produto_maior_vol_estq['Produto']} \nVolume: {produto_maior_vol_estq['Volume total']}")
```

- Qual o produto que possui o maior volume?

```python
df['Volume'] = df['Altura'] * df['Largura'] * df['Profundidade']

id_maior_volume = df['Volume'].idxmax()
produto_maior_volume = df.loc[id_maior_volume]

print("O produto que possui maior volume é:"
      f"\nNome: {produto_maior_volume['Produto']}"
      f"\nVolume: {produto_maior_volume['Volume']}")
```

#### Temos 4 tipos de caixa para embalagem

*   Caixa A: 20x15x8;
*   Caixa B: 30x25x40;
*   Caixa C: 20x25x15;
*   Caixa D: 8x8x4.

Todas unidades estão em centímetros.

#### Crie um dataframe com as seguintes colunas:

*  Nome da caixa;
*  Altura;
*  Largura;
*  Profundidade;
*  Volume total.



```python
from collections import OrderedDict
data = OrderedDict(
{
'Nome' : ["Caixa A", "Caixa B", "Caixa C", "Caixa D"],
'Altura' : [8, 40, 15, 4],
'Largura' : [15, 25, 25, 8],
'Profundidade' : [15, 25, 25, 8],
'Volume total' : [1800, 25000, 9375, 256]
})

df1 = pd.DataFrame(data)
display(df1)
```

- Quantos de nossos produtos nós não podemos embalar com as caixas existentes?

```python
produtos_nao_embalaveis = df[df['Volume'] > df1['Volume total'].max()]

qtd_produtos_nao_embalaveis = produtos_nao_embalaveis.shape[0]

print(f"A quantidade de produtos que não podemos embalar com as caixas é de: {qtd_produtos_nao_embalaveis}")
```

- Quantos de nossos produtos ocupam menos de 50% do volume da menor embalagem?

```python
volume_50 = df1['Volume total'].min() * 0.5

produtos_menos_50 = df[df['Volume'] < volume_50]

qtd_produtos_menos_50 = produtos_menos_50.shape[0]

print(f"A quantidade de produtos que ocupam menos de 50% da menor embalagem é: {qtd_produtos_menos_50}")
```
